### STATISTICAL ANALYSIS

```python
## Hypothesis Test --- One Sample and Two Sample test 
```

```python
## eg. avg men height == avg height women
```

```python
## Write down NUll and the Alternate Hypothesis
# Create sub data from the original based on conditions
# Create samples (Random sampling)
# conclude if we have to go with Null or Alternate Hypothesis

```

```python
new_data = df.loc[(df["Age"] == 1) & (df["Gender"]==1)]
new_data
```

```python
# t-test

# number of samples nearly 30

sample_size = 1000
```

```python
sample= new_data.sample(sample_size)
sample
```

```python
from scipy.stats import ttest_1samp
```

```python
T_value, P_value = ttest_1samp(sample,10000)
```

```python
P_value
```

```python
T_value
```

```python
T_value, P_value = ttest_1samp(sample["Purchase"],10000)
```

```python
P_value
```

```python
if P_value < 0.05:
    print("Alternate Hypothesis: avg purchase made by Men of age 18-25 was not 10000")
else:
    print("Null Hypothesis: avg purchase made by Men of age 18-25 was 10000")
```

```python
data_men = df.loc[(df["Gender"]==1) & (df["Age"]==1)]
data_women =df.loc[(df["Gender"]==0) & (df["Age"]==1)]
```

```python
data_men.shape
```

```python
data_women.shape
```

```python
sample_size = 500
data_men_sample = data_men.sample(sample_size)
data_women_sample = data_women.sample(sample_size)
```

```python
from scipy.stats import ttest_ind

t_stats, p_value = ttest_ind(data_men_sample["Purchase"], data_women_sample["Purchase"], equ_var = True)
```
